```python
import mlxtend
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
import matplotlib.pyplot as plt
```

```python
book = pd.read_csv("book.csv")
```

```python
book
```

```python
book.info()
```

```python
df =pd.get_dummies(book)
df.head()
```

```python
# Apriori algorithm
frequent_itemsets = apriori(df, min_support=0.02,use_colnames=True)
frequent_itemsets.head(50)
                        
```

```python
frequent_itemsets = apriori(df, min_support=0.05,use_colnames=True)
frequent_itemsets.head(50)
```

```python
rules = association_rules(frequent_itemsets, metric="lift",min_threshold=2.0)
rules.head(20)
```

```python
rules.sort_values(by='support', ascending=False)
```

```python
rules[['support','confidence','lift']].hist(figsize=(15,7))
plt.show()
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.show()
```

```python
# Insights 
# CookBks, ChildBks and GeogBks have high support and confidence level therefore they have the strongest association levels
# There are around 250 rows with 0.06 support and 100 rows with 0.05 confience 
```

#  PCA Implementation

```python
# Applying PCA Fit tranform to dataset
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
# PCA components matrix or convariance matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var = np.cumsum(np.round(var,decimals= 4)*100)
var
```

```python
plt.plot(var,color="red")
```

```python
# Final DataFrame
df = pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
df
```

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create Dendrograms
plt.figure(figsize=(11,9))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
# Create Clusters
h_clusters = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
h_clusters
```

```python
y = pd.DataFrame(h_clusters.fit_predict(wine_norm),columns=['clustered'])
y['clustered'].value_counts()
```

```python
# Adding clusters to dataset
wine2=wine.copy()
wine2['clustered']=h_clusters.labels_
wine2
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
# within-cluster sum-of-squares criterion
wcss=[]
for i in range(1,6):
    kmeans = KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values rang vs WCSS to get elbow graph for choosing K (no.of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
```

# Build Cluster Alogorithm Using K-3

```python
# Cluster Algorithm using K=3
clusters3 = KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
wine3['clustered'].value_counts()
```

```python
# Assign clusters to the data set
wine3=wine.copy()
wine3['clustered']=clusters3.labels_
wine3
```
